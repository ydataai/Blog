# Training the synthesizer only on the fraudulent events

For the process of synthesis we shall only consider the fraudulent data.

#### Import the required packages

```python
#ave here YData's lib to be called and used in the Jupyter Notebook
from ydata.connectors.filetype import FileType
from ydata.connectors import LocalConnector

from ydata.synthesizers.regular import RegularSynthesizer
```

#### Reading the fraudulent dataset

```python
conn = LocalConnector()
fraud = conn.read_file(path='outputs/fraud.csv', file_type=FileType.CSV)
```

```python
fraud_synth = RegularSynthesizer()
fraud_synth.fit(fraud)
```

### Saving the trained synthesizer
For the next steps of the pipeline we need the trained synthesizer to generate synthetic fraudulent samples. To enable the synthesizer for the next pipelien steps, it is required to save it as a pickle file.

```python
fraud_synth.save('outputs/synth_model.pkl')
```
